### **Dictionary**

Dictionary in python is `Mutable` and `Unordered` colletion of data values. It is used to store data values like `{key:value}` pair unlike other data structure which hold only single value as an element.

key:value is provided in the dictionary to make it more optimized as we work on real life projects many times our data will be stored in the form of json file and when the data is stored in the form of jason file you need to have a way to quickly access it .

*Dictionary have the following properties :*
- Dictionary is enclosed in {}
- The elements of a dictionary are not accessed using an index but with the help of the key such as dictname["key"].
- Dictionary can be used to save anything such as other lists, dictionaries,....
- Dictionary can be nested.
- A dict is also known as a hash map or associative array.

### Accessing elements in a dictionary

```
Dict = {'name':'Eva', 'age': 20}

print(Dict)
print(Dict.keys())
print(Dict['age'])
print(Dict['name'])
```

```
Dict2 = {'emp': Dict}

print(Dict2)

print(Dict2['emp']['age'])
```

### Adding and changing elements of dictionary

```
car = {'brand': 'ford', 'model': 'Mustang', 'year': '1964'}
print(car)
# adding new element in dictionary
car['color'] = 'black'
print(car)
# changing the exsisting value 
car['model'] = 'figo'
print(car)
```

### Deleting elements from a dictionary 

```
car = {'brand':'ford','model':'Mustang','year':1964,'color':'black'}
print(car)
# deleting element from dictionary 
del car['year']
print(car)
```

### Some built-in methods which can be used on dictionaries 

- **keys()** - Returns a list containing the dictionary's keys


```
Movies = {'Name':'inception',
          'Genre':'sci-fi',
          'Release_yr':2010,
          'Director':' Christopher Nolan'}
print(Movies)
print(Movies.keys())
```

- **values()** - Returns a list of all the values in the dictionary.

```
Movies = {'Name':'inception',
          'Genre':'sci-fi',
          'Release_yr':2010,
          'Director':' Christopher Nolan'}
print(Movies)
print(Movies.values())
```

- **items()** - Returns a list containing a tuple for each key value pair.

```
Movies = {'Name':'inception',
          'Genre':'sci-fi',
          'Release_yr':2010,
          'Director':' Christopher Nolan'}
print(Movies)
print(Movies.items())
```

- **copy()** - Returns a copy of the specified dictionary. With copy, they don't point to the same dictionary anymore, a new copy is created.

```
Movies = {'Name':'inception',
          'Genre':'sci-fi',
          'Release_yr':2010,
          'Director':' Christopher Nolan'}

cpy = Movies.copy()
print(cpy)
```

- **clear()** - Removes all the elements from the dictionary.

```
song = {'Name':'one moment in time',
        'artist':'whitney houston',
        'genre':'pop'}
print(song)

song.clear()
print(song)
```

- **get()** - Returns the value of the specified key.equivalent to `dict['key']`

```
Movies = {'Name':'inception',
          'Genre':'sci-fi',
          'Release_yr':2010,
          'Director':' Christopher Nolan'}
print(Movies)
Movies.get('Name')
```

- **pop()** - Removes the element with the specified key

```
song1 = {'Name':'one moment in time',
        'artist':'whitney houston',
        'genre':'pop',
        'rel_yr':1988}
print(song1)

song1.pop('rel_yr')
print(song1)
```

- **update()** - Updates the dictionary with the specified key-value pairs.

```
car = {'brand':'ford',
       'model':'Mustang',
       'year':1964,
       'color':'black'}
print(car) 

car.update({'color':'white'})
print(car)
```

- **len()** - Returns the number of Key/Value pairs.

```
car = {'brand':'ford',
       'model':'Mustang',
       'year':1964,
       'color':'black'}
print(car)

print(len(car))
```
