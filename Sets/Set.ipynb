## Set

Set is collection of different elements.

**Properties of set**

 - Set elements are `Unordered` - Set items can appear in a different order every time you use them, and cannot be referred to by index or key.
 
 - Sets are `immutable` - we cannot change the items after the set has been created.

 - Set does `not allow duplicate values`.
 
 - Sets elements cannot be accessed by index or key.


### Creating a Set

Set can be defined in two ways:
- By using curly braces `{}`
- By using built-in `set()` function

```
# creating set by using {}
color = {'black','blue','grey','olive','white'}
color
```

```
# creating set with built-in set() function
set1 = set([1,2,2,5,7,6,6,3,3])
set1
```

### Length of set

To determine how many items a set has, use the `len()` method.

```
color = {'black','blue','grey','olive','white'}
print(color)

print(len(color))
```

### set elements - Data Types

Set items can be of any data type. we can check this by using `type()` function.

```
S = {'stars',2.8,8,True}
print(S)
for i in S:
  print("The data type of {} is {}".format(i,type(i)))
```

```
type(S)
```

#### Accessing set elements

*We `cannot access` items in a set by referring to an index or a key.*

But we can loop through the set items using a `for loop`, or ask if a specified value is present in a set, by using the `in` keyword.


```
# printing all elements of the set
color = {'black','blue','grey','olive','white'}

for i in color:
  print(i)
```

```
# checking if element is present in the set
color = {'black','blue','grey','olive','white'}

print('black' in color)
```

### Built-in set Methods

Python has a set of built-in methods that we can use on sets.

- **add()** - Adds an element to the set

*syntax - set.add(value)*

```
color = {'black','blue','grey','olive','white'}
print("Original set : ",color)

color.add('green')
print("set after adding new element : ",color)
```

- **clear -** Removes all the elements from the set.

*syntax : set.clear()*

```
AI  = {'NLP','computer vision','robotics'}
print(AI)

AI.clear()
print(AI)
```

- **copy() -** Returns a copy of the set

*syntax : set.copy()*

```
color = {'black','blue','grey','olive','white'}

new_color = color.copy()
print(new_color)
```

- **pop() -**Removes an random element from the set

*syntax: set.pop()*

```
color = {'black','blue','grey','olive','white'}
print(color)

color.pop()
print(color)
```

- **remove() -** Removes the specified element

*syntax : set.remove(value)*

```
color = {'black','blue','grey','olive','white'}
print(color)

color.remove('white')
print(color)
```

- **union** - Return a set containing the union of sets

we can specify as many sets you want, separated by commas.
If an item is present in more than one set, the result will contain only one appearance of this item.

*syntax : set.union(set1,set2,....)*

```
x = {'google','microsoft','apple'}
print("set x",x)
y = {'facebook','telsa'}
print("set y",y)

z = x.union(y)
print("set union",z)
```

- **intersection() -** Returns a set, that is the intersection of two other sets.(i.e set that contains the similarity between two or more sets)

*syntax : set.intersection(set1)*

```
x = {'google','microsoft','apple'}
print("set x",x)
y = {'facebook','telsa','google'}
print("set y",y)

z = x.intersection(y)
print("set intersection",z)   
```

- **difference() -** Returns a set containing the difference between two or more sets. (i.e The returned set contains items that exist only in the first set, and not in both sets)

*syntax : set.difference(set1)*

```
x = {'google','microsoft','apple'}
print("set x",x)
y = {'facebook','telsa','google'}
print("set y",y)

z = x.difference(y)
print("set difference",z)
```

- **isdisjoint() -** Returns boolean value whether two sets have a intersection or not.

*syntax : set.isdisjoint(set1)*

```
x = {'google','microsoft','apple'}
print("set x",x)
y = {'facebook','telsa','google'}
print("set y",y)

z = x.isdisjoint(y)
print("isdisjoint",z)
```

**issubset() -** returns True if all items in the set exists in the specified set, otherwise it retuns False.

*synax : set.issubse(set1)*

```
x = {"a", "b", "c"}
print("set x",x)

y = {"f", "e", "d", "c", "b", "a"}
print("set y",y)
z = x.issubset(y) 

print("issubset: ",z)
```

**issuperset() -** returns True if all items in the specified set exists in the original set, otherwise it retuns False.


```
x = {"f", "e", "d", "c", "b", "a"}
print("set x",x)

y = {"a", "b", "c"}
print("set y",y)

z = x.issuperset(y) 

print("issuperset",z)
```
